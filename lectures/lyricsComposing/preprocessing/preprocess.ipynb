```python
lyricList = []
p_list = html.xpath('//div[@class="lrc-content"]//p')
for p in p_list:
    lyric = [s for s in p.itertext()]
    lyricList.append(lyric)
    print(lyric)
```

```python
for idx, lyric in enumerate(lyricList):
    filename = "lyrics/lyric-{}.txt".format(idx)
    with open(filename, 'wb') as f:
        for line in lyric:
            # print(line)
            f.write(line.encode("utf-8"))    
    # break
print('finished')
```

#### Build Vocabulary

```python
characters_set = []
for idx, lyric in enumerate(lyricList):
    for line in lyric:
        for charIdx in range(len(line)):
            characters_set.append(line[charIdx])
        characters_set.append('\n')
        
vocabulary = set(characters_set)     
print('vocabulary built')
print('characters_set total size:', len(characters_set))
print('vocabulary set size:', len(set(vocabulary)))
```

```python
vocabulary_dict = {}
vocabulary_search_dict = {}
for idx, w in enumerate(vocabulary):
    vocabulary_dict[w] = idx
    vocabulary_search_dict[idx] = w
```

```python
lyricsNumerical = []
for idx, lyric in enumerate(lyricList):
    lyricNumerical = []
    for line in lyric:
        for charIdx in range(len(line)):
            character = line[charIdx]
            lyricNumerical.append(vocabulary_dict[character])
    print("idx:{},length:{}".format(idx, len(lyricNumerical)))
    lyricsNumerical.append(lyricNumerical)
```

```python
import numpy as np
import scipy.io as sio  

for lyricIdx, lyricItem in enumerate(lyricsNumerical):
    lyricMatrix = np.zeros((len(lyricItem), len(vocabulary)), dtype=np.uint8);
    for idx, word_num in enumerate(lyricItem):
        lyricMatrix[idx, word_num] = 1
        # break
    sio.savemat('lyrics{}.mat'.format(lyricIdx), {'lyric': lyricMatrix})  
```

#### Save vocabulary

```python
with open('vocabulary.txt', 'w') as f:
    for k,v in vocabulary_search_dict.items():
        # print(k,v)
        f.write(str(k)+','+v+'\n')
print('saved vocabulary')
```
